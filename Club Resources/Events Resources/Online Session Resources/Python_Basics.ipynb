### Features of Python

Python is often called a "batteries-included" language. Here are its standout features:

- Simple & Easy to Learn: Python mimics human language. It strips away the complex syntax (like curly braces {} or semi-colons ;) found in languages like C++ or Java.

- Interpreted Language: Python code is executed line-by-line. This makes debugging easy because the program stops immediately when it hits an error, showing you exactly where it failed.

- Dynamically Typed: You donâ€™t need to declare variable types. Python figures out that x = 10 is an integer and name = "Sam" is a string automatically.

- Object-Oriented: It supports concepts like Classes and Objects (Encapsulation, Inheritance, Polymorphism), which helps in building manageable and reusable code.

- Huge Standard Library: Python has massive built-in support for almost anythingâ€”from connecting to web servers (http) to math (math) to file handling (os), minimizing the code you have to write from scratch.

- Portable (Platform Independent): Write once, run anywhere. A Python script written on Windows will run on Linux or Mac without changes (provided the environment is set up).

### Identifier

## Python Operators

Operators are symbols that tell the interpreter to perform specific mathematical or logical manipulations. The values the operator acts on are called **operands**.

For example, in the expression `4 + 5`:

  * `+` is the **Operator**.
  * `4` and `5` are the **Operands**.

Here are the essential categories you need to know.

-----

### 1\. Arithmetic Operators (Math)

These are used to perform common mathematical operations.

| Symbol | Name | Description | Example (`a=10, b=3`) | Result |
| :---: | :--- | :--- | :--- | :--- |
| `+` | Addition | Adds values | `10 + 3` | `13` |
| `-` | Subtraction | Subtracts right from left | `10 - 3` | `7` |
| `*` | Multiplication | Multiplies values | `10 * 3` | `30` |
| `/` | Division (Float) | Divides (always returns float) | `10 / 3` | `3.33...` |
| `//` | Floor Division | Divides and rounds **down** to nearest whole number | `10 // 3` | `3` |
| `%` | Modulus | Returns the **remainder** of division | `10 % 3` | `1` |
| `**` | Exponentiation | Power (Left raised to Right) | `10 ** 3` | `1000` |

> **Note on Floor Division (`//`):** It chops off the decimal. `9 // 2` is `4`, not `4.5`.
> **Note on Modulus (`%`):** Extremely useful for finding even/odd numbers (e.g., `x % 2 == 0` means even).

-----


**Ready for the solution, or shall we move to the next major topic: "Type Casting" (Converting data types)?**

````python
a = 10
b = 3

print("--- Arithmetic ---")
print(f"Addition (10 + 3):       {a + b}")
print(f"Subtraction (10 - 3):    {a - b}")
print(f"Multiplication (10 * 3): {a * b}")
print(f"Division (10 / 3):       {a / b}")   # Returns a float
print(f"Floor Division (10 // 3):{a // b}")  # Returns an int (removes decimal)
print(f"Modulus (10 % 3):        {a % b}")   # Returns the Remainder
print(f"Exponent (10 ** 3):      {a ** b}")  # 10 to the power of 3
print("\n")
````

### 2\. Comparison Operators (Relational)

These compare two values and **always** return a Boolean result (`True` or `False`).

| Symbol | Description | Example | Result |
| :---: | :--- | :--- | :--- |
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `5 > 10` | `False` |
| `<` | Less than | `5 < 10` | `True` |
| `>=` | Greater than or equal to | `5 >= 5` | `True` |
| `<=` | Less than or equal to | `4 <= 5` | `True` |

> **âš ï¸ Crucial Warning:**
>
>   * `=` is for **Assignment** (putting a value in a box).
>   * `==` is for **Comparison** (asking "are these equal?").

-----

````python
x = 5
y = 10
print("--- Comparison ---")
print(f"Is x equal to y? (5 == 10):  {x == y}")
print(f"Is x NOT equal to y? (5 != 10): {x != y}")
print(f"Is x less than y? (5 < 10):  {x < y}")
print("\n")
````


### 3\. Logical Operators

Used to combine conditional statements.

| Operator | Description | Example | Result |
| :---: | :--- | :--- | :--- |
| `and` | Returns True if **both** statements are true | `x < 5 and x < 10` | True only if both met |
| `or` | Returns True if **at least one** statement is true | `x < 5 or x < 4` | True if one is met |
| `not` | Reverses the result | `not(x < 5)` | Flips True to False |

-----

````python
is_adult = True
has_license = False

print("--- Logical ---")
# 'and' needs BOTH to be True
print(f"Can drive? (Adult AND License): {is_adult and has_license}") 

# 'or' needs ONLY ONE to be True
print(f"Needs ticket? (Adult OR License): {is_adult or has_license}") 

# 'not' flips the value
print(f"Is NOT an adult?: {not is_adult}")
print("\n")
````

| `-=` | `x -= 3` | `x = x - 3` |
| `*=` | `x *= 3` | `x = x * 3` |

-----

````python
score = 50
print("--- Assignment ---")
print(f"Initial Score: {score}")

score += 10  # Same as: score = score + 10
print(f"Score after += 10: {score}")

score *= 2   # Same as: score = score * 2
print(f"Score after *= 2: {score}")
print("\n")

````

## Control Flow: Making Decisions

Up until now, our code has been a straight road: Python executes line 1, then line 2, then line 3.

**Control Flow** allows your code to fork. It lets the program ask questions and decide which path to take based on the answer. This is the "brain" of your script.



### 1\. The `if` Statement

This is the most basic decision. "If this condition is True, do this code. If not, skip it."

**Syntax:**

```python
if condition:
    # Do something (Indented!)
```


````python
temperature = 30

if temperature > 25:
    print("It's a hot day!") # This prints
    print("Drink water.")    # This also prints

print("Done.") # This always prints
````

### 2\. The `else` Statement (The Fallback)

The `else` block runs if the `if` condition was **False**. It catches everything else.

**Syntax:**

```python
if condition:
    # path A
else:
    # path B
```


````python
age = 16

if age >= 18:
    print("Access Granted")
else:
    print("Access Denied") # This runs because 16 is not >= 18
````

### 3\. The `elif` Statement (Multiple Checks)

Short for "Else If". This allows you to check multiple conditions in a specific order. Python stops at the **first** one that is True.

````python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B") # Python stops here!
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
````

> **Critical Logic Note:** Order matters\! If I swapped the order and put `elif score >= 70` first, it would have printed "Grade: C" because 85 is greater than 70. Always order your checks from most specific to least specific.


### 4\. Nested Ifs (Decisions inside Decisions)

You can put an `if` statement inside another `if` statement. This is useful for filtering data in layers.


````python
number = 10

if number > 0:
    print("Positive number")
    
    # Inner check
    if number % 2 == 0:
        print("It is Even")
    else:
        print("It is Odd")
        
else:
    print("Negative number")
````

### 5\. Short Hand (Ternary Operators)

For very simple `if/else` logic, you can write it in one line. This is considered "Pythonic" for assigning variables.

**Syntax:** `value_if_true` **if** `condition` **else** `value_if_false`


````python
age = 20
status = "Adult" if age >= 18 else "Minor"

print(status) # Output: Adult
````

## Loops: Repetition

Loops allow you to execute a block of code multiple times. This is essential for automationâ€”whether you are processing 10 items or 10 million.

There are two main types of loops in Python:

1.  **`for` loops:** Used when you know *how many times* you want to loop (or have a list of things).
2.  **`while` loops:** Used when you want to loop *until a condition changes*.

-----


### 1\. The `for` Loop

The `for` loop iterates over a **sequence** (like a list, a string, or a range of numbers).

**Syntax:**

```python
for item in sequence:
    # Do something with 'item'
```


#### A. Looping through a List


````python
fruits = ["Apple", "Banana", "Cherry"]

for fruit in fruits:
    print(f"I like {fruit}")
# Prints:
# I like Apple
# I like Banana
# I like Cherry
````




#### B. Looping through a String

It goes character by character.


````python
for char in "Python":
    print(char) 
````

#### C. The `range()` Function

This is how you loop a specific number of times.

  * `range(5)`: 0, 1, 2, 3, 4 (Starts at 0, stops *before* 5)
  * `range(1, 6)`: 1, 2, 3, 4, 5 (Starts at 1, stops *before* 6)
  * `range(0, 10, 2)`: 0, 2, 4, 6, 8 (Step size of 2)

<!-- end list -->



````python
for i in range(1, 4):
    print(f"Lap {i}")
````



### 2\. The `while` Loop

The `while` loop keeps running **as long as** a condition is `True`.

**Syntax:**

```python
while condition:
    # Do something
    # IMPORTANT: Update condition to avoid infinite loop
```

**Example:**

```python
count = 5

while count > 0:
    print(f"Countdown: {count}")
    count = count - 1  # Decrement logic

print("Blast off!")
```

> **âš ï¸ Danger Zone: The Infinite Loop**
> If you forget `count = count - 1`, the loop checks `5 > 0` forever. Your program will freeze (Press `Ctrl+C` in the terminal to kill it).

-----


### 3\. Loop Control: `break` and `continue`

Sometimes you need to alter the flow of the loop manually.

  * **`break`**: Stops the loop *immediately* and exits.
  * **`continue`**: Skips the *current* iteration and jumps back to the start of the loop.



````python
numbers = [1, 5, 8, 12, 7]

for num in numbers:
    if num == 12:
        print("Found 12! Stopping.")
        break # Exits the loop entirely
    print(f"Checking {num}...")
````

````python

for i in range(5):
    if i == 2:
        continue # Skips printing 2
    print(i)
# Prints: 0, 1, 3, 4

````
